```python
GET은 요청 바디를 사용하지 않는 게 표준이고,
POST는 요청 바디를 사용하는 게 기본 설계입니다.

📦 이유 설명
🔹 GET 메서드의 목적
데이터 조회용

getdms URL만으로 모든 정보를 표현해야 함  ///중요

예:

GET /user?id=123
📛 HTTP 표준(RFC 7231) 에 따르면:

“GET 메서드는 바디를 포함하지 않아야 한다.”
(Some servers ignore the body even if it exists.)

➡ 서버나 프록시가 GET 요청의 body를 무시할 수 있으므로, 일관성과 호환성 문제 생김

🔸 POST 메서드의 목적
서버에 데이터 전달/생성

바디에 원하는 만큼 데이터 담을 수 있음

로그인, 회원가입, 폼 제출 등은 모두 POST로 처리
```

```python
userSchema.methods.toJSON = function () {
  const obj = this._doc;
  delete obj.password;
  return obj;
}; // 유저 스키마 중 _doc에서 패스워드 빼고 반환 (처음에 스키마 전체 this로 출력해보고 _doc에 우리가 찾는 정보들만 있는거 확인하고 _doc에서 obj에 넣기)

userSchema.methods.toJSON은 Mongoose가 .toJSON()을 호출할 때 자동으로 실행되는 커스텀 오버라이드 메서드입니다.

즉, 아래처럼 명시적으로 .toJSON()을 호출하거나,
자동으로 JSON 직렬화될 때(예: res.json(user)), 자동으로 실행
##toJSON() 메서드는 "JSON으로 변환되는 순간"에 자동 호출
```
