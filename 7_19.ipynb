```python
req.body에서 body는
클라이언트(예: 프론트엔드, Postman 등)가 서버로 HTTP 요청을 보낼 때 함께 담아 보내는 데이터(본문)를 의미합니다.
자세한 설명
HTTP 요청에는 여러 부분이 있습니다:
헤더(header): 요청 정보(타입, 인증 등)
쿼리(query): 주소 뒤에 붙는 데이터 (?name=홍길동)
파라미터(params): 주소 경로에 포함된 데이터 (/users/1)
바디(body): 요청의 본문에 담기는 데이터 (주로 POST, PUT, PATCH에서 사용)
req.body란?
req: 요청 객체(request object)
body: 요청의 본문(Body)에 담긴 데이터
```

```python
body는 즉, 요청의 본문을 저장하고 읽어 오는 CRUD의 수행 데이터를 받아오기 위해 필요한거지?

네, 정확하게 이해하셨습니다!
body는
요청의 본문(Body)에 담긴 데이터를 저장하고 읽어오는 역할을 합니다.
주로 CRUD(생성, 수정 등) 작업에서 클라이언트가 보낸 데이터를 받아오기 위해 필요합니다.
예시로 정리
Create(생성):
POST 요청의 body에 새로 만들 데이터(예: 할 일 내용)를 담아 보냄
→ 서버는 req.body에서 꺼내서 DB에 저장
Update(수정):
PUT/PATCH 요청의 body에 수정할 데이터(예: 완료 여부 등)를 담아 보냄
→ 서버는 req.body에서 꺼내서 DB에 반영
결론
body는 요청의 본문에 담긴 데이터를 받아오기 위한 것
CRUD 작업에서 클라이언트가 보낸 실제 데이터(내용, 상태 등)를 읽어오는 데 꼭 필요
```

```python
클라이언트가 JSON 형식이 아닌 다른 방식(예: 폼 데이터, 텍스트 등)으로 데이터를 보내면
req.body에 원하는 데이터가 제대로 담기지 않을 수 있습니다.
이 경우에는 추출 또는 파싱 작업이 추가로 필요할 수 있습니다.
자세한 설명
1. JSON 형식으로 보낼 때
Content-Type: application/json 헤더와 함께 JSON 데이터로 요청을 보내면
express.json()(또는 body-parser의 json 미들웨어)이 자동으로 파싱해서
req.body에 객체로 담아줍니다.
별도의 추출/파싱 작업이 필요 없습니다.
2. 폼 데이터(x-www-form-urlencoded)로 보낼 때
Content-Type: application/x-www-form-urlencoded로 보내면
express.urlencoded() 미들웨어가 필요합니다.
이 미들웨어가 없으면 req.body가 비어 있거나 undefined가 됩니다.
3. 텍스트, 멀티파트 등 기타 형식
해당 형식에 맞는 미들웨어가 필요합니다.
없으면 직접 파싱하거나, 데이터를 수동으로 추출해야 할 수 있습니다.
```

```python
. CORS (Cross-Origin Resource Sharing)
백엔드 서버에서 "어떤 출처(origin, 즉 도메인/포트/프로토콜 조합)에서 오는 요청을 허용할지"를 설정합니다.

즉, 내 API 서버에 접근할 수 있는 프론트엔드의 주소(Origin)를 허락하는 것.
브라우저가 백엔드에 요청을 보낼 때, 백엔드가 허락한 Origin이 아니면 브라우저가 요청을 차단합니다.
2. Firebase의 승인된 도메인
Firebase Authentication, Hosting, Storage 등에서 "이 서비스에 접근할 수 있는 웹사이트의 도메인"을 등록합니다.
예시:
인증(로그인) 리디렉션 허용 도메인
호스팅 허용 도메인
Firebase 서비스에 접근할 수 있는 프론트엔드의 주소를 미리 등록하는 것.
등록되지 않은 도메인에서 접근하면 Firebase가 자체적으로 차단합니다.
공통점
둘 다 "어떤 프론트엔드(도메인)에서 접근을 허용할지"를 설정합니다.
보안상의 이유로, 허용된 도메인만 접근할 수 있도록 제한합니다.
차이점
CORS는 "브라우저와 백엔드 API 서버" 사이의 통신을 제어합니다.
Firebase 승인 도메인은 "Firebase 서비스(인증, 호스팅 등)와 프론트엔드" 사이의 접근을 제어합니다.
CORS는 HTTP 레벨에서, Firebase 도메인 등록은 서비스 레벨에서 동작합니다.
```

```python
상속:
"이 클래스가 다른 클래스의 기능을 확장해서 써야 할 때"
클래스 간 계층 구조(부모-자식 관계)가 필요할 때 사용
import:
"이 파일(모듈)에서 다른 파일(모듈)의 코드가 필요할 때"
파일/모듈 간 코드 공유가 필요할 때 사용
예시로 쉽게 구분
상속:
"나는 동물의 모든 특성을 갖고, 거기에 내 것만 추가할래!" → 상속
import:
"저 파일에 있는 함수/클래스/변수를 여기서도 쓸래!" → import
```

```python
import는 상속과 다르게 "그 파일(모듈)에서만" 적용됩니다.
자세히 설명
import는
다른 파일(모듈)에 정의된 함수, 변수, 클래스 등을 "가져와서"
import한 파일 안에서만 사용할 수 있게 해줍니다.
import한다고 해서, 그 기능이 내 코드에 "붙어서 상속되는 것"이 아니라
"필요할 때만 불러와서 쓰는 것"입니다.

import: 가져온 기능은 import한 "그 파일에서만" 사용 가능, 상속처럼 내 것이 되는 게 아님
상속: 자식 클래스가 부모의 기능을 "내 것처럼" 가짐
```

```python
props는 React에서 컴포넌트 간에 데이터를 전달할 때 사용하는 "속성(properties)"입니다.
props란?
props는 "properties"의 줄임말입니다.
부모 컴포넌트가 자식 컴포넌트에게 값을 전달할 때 사용합니다.
자식 컴포넌트는 props를 읽기 전용(read-only)으로 받아서 사용할 수 있습니다.

props는 부모 → 자식 컴포넌트로 값을 전달하는 방법
자식 컴포넌트는 props를 받아서 화면에 표시하거나, 내부 로직에 사용할 수 있음
props는 읽기 전용 (자식이 직접 수정할 수 없음)
```

```python
1. todolist 요소마다 각 정보를 받아옴
todoList는 할 일 객체들의 배열입니다.
각 객체는 예를 들어 { _id: "123", task: "공부하기", isComplete: false }와 같은 구조입니다.
2. map으로 각 요소를 반복(mapping)
todoList.map((item, index) => ( ... ))
→ 배열의 각 요소(item)를 반복해서,
→ 각 item을 TodoItem 컴포넌트에 props로 전달합니다.
3. 각 상황에 맞는 함수 호출
각 TodoItem은
item(할 일 정보)
deleteItem(삭제 함수)
toggleComplete(완료/미완료 토글 함수)
를 props로 받아서,
버튼 클릭 등 이벤트가 발생하면
→ 해당 item의 id를 인자로 하여
→ 알맞은 함수를 호출합니다.
```

```python
아래에서 map과 이벤트, 그리고 map의 호출 시점에 대해 쉽게 설명드릴게요.
1. map으로 (item, index)를 받아서 각 함수에 넣어주는 것
todoList.map((item, index) => ( ... ))
→ map 함수는 배열의 각 요소(item)와 그 인덱스(index)를 받아서
→ 그걸로 컴포넌트(여기서는 <TodoItem />)를 만듭니다.
각 TodoItem에
item(할 일 정보)
deleteItem, toggleComplete(함수)
를 props로 전달합니다.
이벤트(버튼 클릭 등)가 발생할 때
→ 버튼의 onClick에서 item._id 등 해당 item의 정보를 함수에 넣어서 호출합니다.
2. map이 언제 호출되는가?
map 함수는 "렌더링(render)" 시점에 호출됩니다.
즉,
컴포넌트가 처음 화면에 나타날 때
또는 todoList가 바뀔 때(추가/삭제/수정 등)
→ React가 화면을 다시 그릴 때(map이 다시 실행됨)
map 함수 자체는 이벤트가 발생할 때마다 호출되는 게 아니라,
화면에 할 일 목록을 그릴 때마다 호출됩니다.
예시 흐름
초기 렌더링
todoList에 3개의 할 일이 있으면, map이 3번 반복되어 3개의 TodoItem이 만들어짐
할 일 추가/삭제/수정
todoList가 바뀌면, React가 다시 렌더링 → map이 다시 실행됨
버튼 클릭(이벤트 발생)
map이 아니라, 해당 TodoItem의 onClick 함수가 실행됨
이때, map에서 props로 넘겨준 item의 정보가 함수에 전달됨
정리
map은 배열을 화면에 그릴 때(렌더링 시) 반복해서 컴포넌트를 만듦
이벤트(버튼 클릭 등)는 각 컴포넌트에서 발생하며, 그때 item 정보를 함수에 넘김
map은 이벤트 발생 시마다 호출되는 게 아니라, 화면이 다시 그려질 때 호출됨
```

```python
1. map의 역할
todoList.map(...)은 할 일 배열을 화면에 그리기 위해 반복해서 <TodoItem /> 컴포넌트를 생성합니다.

각 item과 index를 받아서 컴포넌트에 props로 전달합니다.

2. 이벤트 처리
각 <TodoItem /> 안에는 삭제/완료 버튼이 있고, 버튼 클릭 시 onClick으로 전달받은 deleteItem, toggleComplete 함수가 실행됩니다.

이때 item._id 같은 정보가 함께 함수에 전달됩니다.

3. map의 호출 시점
map은 화면이 렌더링될 때 호출됩니다.

즉, 컴포넌트가 처음 나타날 때

또는 todoList 배열이 변경될 때(추가, 삭제 등)

버튼 클릭 같은 이벤트가 map을 호출하지는 않습니다.

🔁 전체 흐름 요약
map: 렌더링 시 실행 → 여러 개의 <TodoItem /> 생성

이벤트: 각 <TodoItem /> 안에서 발생 → 전달된 함수 실행

map은 이벤트 발생 시가 아니라 화면이 바뀔 때 다시 실행됨

```

```python
✅ 흐름 정리
✅ "끝남" 버튼 클릭 → toggleComplete 실행

✅ todoList 상태를 바꿔서 setTodoList(...) 호출

✅ 상태가 바뀌었기 때문에 컴포넌트 리렌더링 발생

✅ 리렌더링 시 map(...)이 다시 실행되어 변경된 todoList로 화면 다시 그림

📌 핵심 포인트
버튼 클릭 = 이벤트 발생

map() = 렌더링 시 실행되는 배열 반복 로직

상태 변경 (setTodoList)이 있어야 map()이 다시 실행됨

따라서, toggleComplete이 내부에서 상태를 변경하면 리렌더링 → map() 재실행
```

```python
📌 핵심 포인트
버튼 클릭 = 이벤트 발생
이때만 toggleComplete(혹은 deleteItem 등)이 실행됨
map() = 렌더링 시 실행되는 배열 반복 로직
이벤트 발생 시마다 map이 실행되는 게 아니라,
상태가 바뀌어서 리렌더링이 일어날 때 map이 다시 실행됨
상태 변경(setTodoList)이 있어야 map()이 다시 실행됨
상태가 바뀌지 않으면 map도 다시 실행되지 않음
따라서, toggleComplete이 내부에서 상태를 변경하면
리렌더링 → map() 재실행 → 화면이 최신 상태로 다시 그림
```

```python
**브랜치(branch)**는 독립적으로 작업을 진행할 수 있게 해주는 Git의 기능이에요.

예를 들어, 여러 명이 동시에 기능을 개발하거나 실험할 수 있도록 작업 공간을 분리해주는 역할을 합니다.

🔹 master 브랜치란?
Git 저장소를 처음 만들면 자동으로 생성되는 기본 브랜치입니다.

보통은 최종 배포본(=정식 버전) 이 올라가는 곳이에요.

최근에는 master 대신 main 이라는 이름을 기본으로 쓰기도 해요 (GitHub에서는 main이 기본).

🔹 왜 master를 사용했을까?
전통적으로 "중심(branch의 주축)"이라는 의미로 master가 사용됐습니다.

다만 최근에는 포괄적 언어 사용을 위해 GitHub, GitLab 등에서는 main을 기본 브랜치로 변경
```

```python
예전(2020년 이전)에는 git의 기본 브랜치 이름이 master였습니다.
최근에는 git과 GitHub 등에서 기본 브랜치 이름을 main으로 바꾸는 추세입니다.
정리
복사한 코드(레포)가 예전 방식이면 기본 브랜치가 master일 가능성이 높습니다.
그래서 git push origin master가 맞는 명령어입니다.
만약 새로 만든 저장소라면 기본 브랜치가 main일 수도 있으니,
항상 git branch로 내 브랜치 이름을 확인하는 습관이 좋습니다.
```

```python
http와 https의 차이는 뭘까🤨
둘다 인터넷에서 데이터를 전송하기위해사용되는 프로토콜(전세계적으로 약속된 규칙) 이다. 하지만 http는 보안기능이 없어 데이터가 암호화되지 않은채로 전송이되고 https는 데이터를 보호하기위해 암호화와 인증절차가 있다.

http처럼 암호화가 안되어있으면 중간에 데이터를 가로채기 쉽다. https는 암호화가 되어있어 가로채도 읽기가 어렵다.
https는 아무나와 통신하지 않는다. 내가 통신하고자 하는 상대가 누구인지 인증과정을 거쳐야한다. (영상에서도 봤다시피 https는 http를 받아주려하지 않는다)
```
